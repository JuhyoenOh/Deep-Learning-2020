# 과적합 피하기 - 초음파광물 사례

```python
import pandas as pd

df = pd.read_csv('dataset/sonar.csv', header=None)
df.head()
```

```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
```

```python
# seed 값 설정
seed = 2020
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
dataset = df.values
X = np.array(dataset[:,0:60], dtype=np.float64)
Y_obj = dataset[:,60]
```

```python
type(Y_obj)
```

```python
# 문자열 변환
e = LabelEncoder().fit(Y_obj)
Y = np.array(e.transform(Y_obj), dtype=np.float64)
```

```python
Y_obj[:5], Y[:5]
```

```python
# 모델 설정
model = Sequential([
    Dense(24, input_dim=60, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# 모델 컴파일 
model.compile(loss='binary_crossentropy',   # mean_squared_error
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=5)

# 에포크(epoch) 횟수가 늘어나면 과적합이 발생을 한다.
```

```python
# 결과 출력 
print("\n Accuracy: %.4f" % (model.evaluate(X, Y, verbose=2)[1]))
```
