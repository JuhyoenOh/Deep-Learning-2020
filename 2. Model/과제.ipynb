# 이중분류

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt
%matplotlib inline

from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
bc = load_breast_cancer()

bc_df = pd.DataFrame(data=bc.data)
bc_target = pd.DataFrame(data=bc['target'])

dataset_1 = bc_df.values
dataset_2 = bc_target.values 

X = dataset_1[:,:]
Y = dataset_2[:,:1]
```

```python
model = Sequential([
    Dense(30, input_dim=30, activation='relu'),
    Dense(12, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
]) 
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# 자동 중단 설정
early_stopping_callback = EarlyStopping(monitor='loss', patience=100)
```

```python
history = model.fit(X, Y, epochs=1500, batch_size=200)# epochs 늘리면 됨
```

```python
# 결과 출력 
print("\n Accuracy: %.4f" % (model.evaluate(X, Y, verbose=2)[1]))
```

```python
y_vloss=history.history['loss']
```

```python
y_acc=history.history['accuracy']
```

```python
x_len = np.arange(len(y_acc))
plt.figure(figsize=(12,8))
plt.plot(x_len, y_vloss, "o", c="red", markersize=2)
plt.plot(x_len, y_acc, "o", c="blue", markersize=2)
plt.ylim((0,1))
plt.show()
```

```python
model.save('model/my_model-1.h5')
```

```python
del model 
```

# 다중분류

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

tf.__version__
```

```python
bc = load_breast_cancer()
```

```python
bc
```

```python
print(bc['target'])
```

```python
seed = 2020
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
bc_target# 타겟 확인해본거
```

```python
bc_df = pd.DataFrame(data=bc.data) # 그냥 데이터 만 집어넣은거
bc_target = pd.DataFrame(data=bc['target']) # 타겟만 빼서 데이타 프레임 집어 넣은거

dataset_1 = bc_df.values # X값이 데이터 전부 들어가야 하니까 bc_df 넣음
dataset_2 = bc_target.values # 이거는 다른놈 넣음

X = dataset_1[:,:] # 각각은 X Y에 넣는데 X는 모든 데이타 들어가면 되니까 걍 모두 들어가라고 값을 안넣음
Y = dataset_2[:,:1] # 이놈은 타겟만 들어가면 되니까 1줄만 있고 걍 이거 다 들어감 그리고 스필릿하고 나머지는 다 걍 박아버림
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size =0.2,shuffle=True, stratify=Y,        random_state=seed)
```

```python
model = Sequential()
model.add(Dense(24, input_shape=(30,),  activation ='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
#제가 알기로는 덴스는 걍 은닉 어쩌구라서 상관 없는듯
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 자동 중단 설정
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=100)
```

```python
history = model.fit(X_train, Y_train, validation_split = 0.2 ,epochs=1000, batch_size=100) ## 
```

```python
print("\n Accuracy: %.4f" % (model.evaluate(X_test, Y_test, verbose=2))[1])
```

```python
y_vloss=history.history['val_loss']
```

```python
y_acc=history.history['accuracy']
```

```python
x_len = np.arange(len(y_acc))
plt.figure(figsize=(12,8))
plt.plot(x_len, y_vloss, "o", c="red", markersize=2)
plt.plot(x_len, y_acc, "o", c="blue", markersize=2)
plt.ylim((0,1))
plt.show()
```

```python
y_vacc=history.history['val_accuracy']
```

```python
y_loss=history.history['loss']
```

```python
x_len = np.arange(len(y_acc))
plt.figure(figsize=(12,8))
plt.plot(x_len, y_loss, "o", c="red", markersize=1)
plt.plot(x_len, y_vacc, "o", c="blue", markersize=1)
plt.plot(x_len, y_vloss, "o", c="orange", markersize=1)
plt.plot(x_len, y_acc, "o", c="green", markersize=1)
plt.ylim((0,1))
plt.show()
```
